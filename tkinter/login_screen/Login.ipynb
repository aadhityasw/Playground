```python
import tkinter as tk
import tkinter.messagebox
```

```python
class Login :
    def __init__(self, loginscr) :
        tk.Label(loginscr, text = "User Name").grid(row = 0)
        self.a=tk.Entry(loginscr)
        self.a.grid(row=0, column=1)
        tk.Label(loginscr, text="Password").grid(row=1)
        self.b=tk.Entry(loginscr, show='*')
        self.b.grid(row=1, column=1)
        self.but=tk.Button(loginscr, text="Submit")
        self.main()
    def on_press(self,but) :
        user_name = self.a.get()
        password = self.b.get()
        print("User Name :-",user_name)
        print("Password :-",password)
    def on_left_press(self,but) :
        error = tk.Tk()
        tk.Label(error, text = "No left clicks allowed").pack()
        error.mainloop()
    def main(self) :
        self.but.grid(row=2, column =1)
        self.but.bind("<Button-1>",self.on_press)
        self.but.bind("<Button-2>",self.on_left_press)
        
        
loginscr = tk.Tk()
loginscr.title("Login")
log = Login(loginscr)
loginscr.mainloop()
```
